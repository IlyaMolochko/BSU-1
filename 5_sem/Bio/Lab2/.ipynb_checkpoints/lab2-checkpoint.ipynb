Импортируем необходимые библиотеки.

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

Зададим параметры

```python
t0 = 23
N0 = 59
r = 1.58
k = 77
step = 0.1
```

Посчитаем числинность популяции в зависимости от времени

```python
t = np.linspace(t0, 2 * t0, int(t0 / step + 1e-9) + 1)
N1 = N0 * np.exp(r * (t - t0))
N2 = 77 * 59 * np.exp(1.58 * (t - 23)) / (77 + 59 * (np.exp(1.58 * (t - 23)) - 1))
N3 = np.zeros(10 * t0 + 1)
N3[0] = N0
for i in range(1, N3.shape[0]):
    N3[i] = N3[i - 1] + step * r * N3[i - 1] * (1 - N3[i - 1] / k)
```

```python
data = pd.DataFrame(np.concatenate((t.reshape(t.shape[0], 1), 
                                    N1.reshape(t.shape[0], 1), 
                                    N2.reshape(t.shape[0], 1), 
                                    N3.reshape(t.shape[0], 1)), axis=1),
                    columns=['t', 'N(t)_своб', 
                             'N(t)_огр', 'n(t)_Эйлер'])
```

Вывыедем всё в виде таблицы

```python
data
```

Нарисуем графики

```python
plt.figure(figsize=(8, 8))
plt.plot(t, N1, label='Свободный')
plt.plot(t, N2, label='Ограниченный')
# plt.plot(t, N3, label='Эйлер')
plt.ylim([40, 150])
plt.legend()
plt.grid()
plt.show()
```
