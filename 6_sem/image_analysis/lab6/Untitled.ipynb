```python
from skimage import io as skio
import numpy as np

from matplotlib import pyplot as plt
```

```python
img1 = skio.imread('1016-cat.jpg', as_grey=True)
img1 = (img1 * 256).astype(int)

with open('cman_uint16.raw', 'rb') as fin:
    img2 = np.fromfile(fin,dtype=np.uint16)
img2 = img2.reshape(256, 256)

    
with open('Pelvis_380_300_31_16bit_roi.raw', 'rb') as fin:
    img3 = np.fromfile(fin,dtype=np.uint16)
img3 = img3.reshape(380, 300, 31)
```

```python
plt.imshow(img2)
```

```python
plt.hist(img1.reshape(-1), bins=256, range=(0, 256))
plt.show()
```
